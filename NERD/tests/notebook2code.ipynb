```python
def notebook2py(nb, pysrc):
    with open(nb) as inp:
        data = json.load(inp)
        
    cells = data['cells']
    
    with open(pysrc, 'w') as out:
        for cell in cells:
            src = cell['source']
            cell_type = cell['cell_type']
            if cell_type == 'code':
                if len(src) == 0:
                    continue
                if src[0].startswith('#code'):
                    for line in src[1:]:
                        out.write(line)
            
            out.write('\n\n')
```

```python
notebook2py('UserManagement.ipynb', '../nerdlogin.py')
```
